```python
import pickle
import pandas
import numpy as np
pandas.set_option('display.max_rows', None)
```

## Disctionary 1 = plate 1

```python
subs_dict = pandas.read_pickle('d1.pkl')
```

```python
subs_dict
```

```python
subs_dict['A1']['Pyr']
```

```python
rates = pandas.read_excel(r"C:\Users\Francel\Pipetting robot scripts\output_plate1_pH8.xlsx", sheet_name='plate1',
                         index_col=0)
```

```python
rates
```

```python
rates['NADH'] = 0.0
rates['Pyr'] = 0.0
```

```python
rates.head()
```

```python
rates.loc['A1']['NADH']
```

```python
rates.index
```

```python
for key in subs_dict:
    if key in rates.index:
        rates.loc[key]['NADH'] = subs_dict[key]['NADH']['conc']
        rates.loc[key]['Pyr'] = subs_dict[key]['Pyr']['conc']
```

```python
rates
```

```python
writer = pandas.ExcelWriter('output_plate1.xlsx')
rates.to_excel(writer, sheet_name='exp')
writer.save()
```
